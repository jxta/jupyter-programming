## 問 7-1 $x^3$のグラフを描く

 $x^3$の値を$x$が-10から10まで計算し、折れ線グラフにしなさい
 
 手順は次の通りとしなさい。

1. $x^3$を計算する関数x3()を定義する
1. $x$に-10から10までのリストを用意する
1. $x$ にx3()を適用し、計算結果をリスト$y$に代入する
1. $x$, $y$とplt.plot()を使ってグラフを描画する

```python
def x3(n):
    return (n**3)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = list(map(x3,x))
plt.plot(x,y)
```

## 問 7-2 階乗のプロット

階乗(fact)を$x$が1から10まで計算し、plt.plot()を使ってグラフ化しなさい。

手順は次の通りとしなさい。

1. 階乗を計算するfact()を定義する

1. $x$に1から20までのリストを用意する

1. $x$ にfact()を適用し、計算結果をリストyに代入する

1. $x$, $y$とplt.plot()を使ってグラフを描画する

```python
def fact(n):
    if (n == 0):
        return 1
    else:
        return n * fact(n-1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = list(map(fact,x))
plt.plot(x,y)
```

## 問 7-3 フィボナッチ数のプロット

フィボナッチ数列を求める関数を定義し、1から20までの値をグラフにプロットしなさい。


1. フィボナッチ数列を計算するfib()を定義する。

1. $x$に1から20までのリストを用意する

1. $x$にfib()を適用し、計算結果をリスト$y$に代入する

1. $x$, $y$とplt.plot()を使ってグラフを描画する



```python
def fib(x):
    if (x == 0):
        return 0
    if (x == 1):
        return 1
    return fib(x-1) + fib(x-2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = list(range(1, 21)) # 1から20までのリスト
y =list(map(fib, x)) # リストの数字の階乗を計算する
plt.plot(x,y) # プロットする
```
