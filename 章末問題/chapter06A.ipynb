## 6-4

```python
def isPrime(n):
    if (n < 2):
        return False
    for i in range(2,n):
        if (n %i ) == 0:
            return False
    return True
```

```python
for i in range(1,101):
    if isPrime(i):
        print (i, end=" ")
```

## 6-5

```python
def divisor(n):
    for i in range(1, n+1):
        if (n % i) == 0:
            print (i, end=" ")
```

```python
divisor(60)
```

```python
def n_divisor(n):
    cnt = 0
    for i in range(1, n+1):
        if (n % i) == 0:
            cnt += 1
    return (cnt)
```

```python
n_divisor(60)
```

```python
def factorization(n):
    if isPrime(n):
        print(n, end=" ")
        return
    for i in range(2, n):
        if (n % i)== 0:
            print(i, end=" ")
            factorization(int(n/i))
            break
```

```python
factorization(30)
```

```python
for i in range(1001, 1010):
    print(i, end=" ")
    factorization(i)
    print()
```

## 6-6

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
```

```python
for i in range(21):
    print ("fact(", i, ")=", fact(i))
```

## 6-7

```python
def fib(x):
    if x == 0:
        return 0
    if x == 1:
        return 1
    return fib(x-1) + fib(x-2)
```

```python
for i in range(21):
    print ("fib(", i, ")=", fib(i))
```

## 6-8

```python
def fibcall(x):
    if x == 0:
        return 1
    if x == 1:
        return 1
    return 1 +fibcall(x-1) + fibcall(x-2)
```

```python
for i in range(31):
    print ("fibcall(", i, ")=", fibcall(i))
```
