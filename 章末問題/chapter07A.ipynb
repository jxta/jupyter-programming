## 7-4

```python
%matplotlib inline
import matplotlib.pyplot as plt

def fact(n):
    if  n == 0:
        return 1
    else:
        return n * fact(n-1)

x = list(range(1, 11))
y = list(map(fact,x))
plt.plot(x,y)
```

## 7-5

```python
%matplotlib inline
import matplotlib.pyplot as plt

def fib(x):
    if x == 0:
        return 0
    if x == 1:
        return 1
    return fib(x-1) + fib(x-2)

x = list(range(1, 21)) # 1から20までのリスト
y =list(map(fib, x)) # リストの数字の階乗を計算する
plt.plot(x,y) # プロットする
```
