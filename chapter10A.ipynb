## 問10-1 ヒストグラム

1次元の酔歩問題で、1000ステップ実行した場合の位置のヒストグラムを作成しなさい。ただし、ヒストグラムの作成にはplt.hist()を使用しなさい。

```python
import random
import math
import matplotlib.pyplot as plt

def next_step():
    if (random.random() > .5):
        return 1
    else:
        return -1
```

```python
pos = 0
pos_history = [] # 歩みを記録するリスト
for i in range(1000):
    pos += next_step()
    pos_history.append(pos) # リストに位置を記録する

plt.hist(pos_history,  rwidth = 0.9)
plt.show()
```

## 問10-2 1次元の酔歩（連続移動）

1次元の酔歩問題で、歩く距離が-1 から+1の範囲でランダムに連続的に変化する場合のプログラムを作成しなさい。

```python
def next_stepC():
    return random.random() * 2 - 1
```

```python
pos = 0
pos_history = [] # 歩みを記録するリスト
for i in range(1000):
    pos += next_stepC()
    pos_history.append(pos) # リストに位置を記録する

plt.plot(pos_history)
plt.show()
```

## 問10-4 移動距離が関数で与えられる場合の一次元の酔歩

1次元の酔歩問題で、移動距離が次の式で与えられる場合のシミュレーションを実施して、移動位置の履歴とヒストグラムを表示しなさい。

座標Xとした場合、次の移動距離dは次の式で与えられる

$$
\left\{
\begin{array}{l}
d(x) = - \frac{1}{x} * random() ( |x| < 0.1 ) \\
d(x) = 2- random() (それ以外)
\end{array}
\right.
$$

ただし、random()は0以上1未満の乱数とする。

```python
def next_stepS(x):
    if (abs(x) < .1):
        return 2 - random.random()
    else:
        return - random.random() * 2 / x  
```

```python
pos = 0
pos_history = [] # 歩みを記録するリスト
for i in range(1000):
    pos += next_stepS(pos)
    pos_history.append(pos) # リストに位置を記録する

plt.plot(pos_history)
plt.show() 
```
